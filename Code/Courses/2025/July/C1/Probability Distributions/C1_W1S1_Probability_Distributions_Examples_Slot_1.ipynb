# **Discrete vs Continuous Random Variables**

A **random variable** is a variable that takes on numerical values determined by the outcome of a random process.

Random variables are classified into two main types: Discrete and Continuous
<br><br>


| Characteristic                | Discrete Random Variable                | Continuous Random Variable                   |
|------------------------------|------------------------------------------|-----------------------------------------------|
| Type of values               | Countable                               | Measurable                                     |
| Number of possible outcomes  | Finite or countably infinite             | Uncountably infinite (e.g., real numbers)     |
| Probability function         | PMF (Probability Mass Function)         | PDF (Probability Density Function)            |
| Probability of a specific value | Non-zero (e.g., $\mathrm{P(X=3) = 0.2}$)        | Zero ($\mathrm{P(X=3.5) = 0}$), area under PDF is used   |


<br><br>
**Real-World Examples**

- **Website Hits per Hour (Discrete):**  
  You can count how many users visited the site — 0, 1, 2, 3, etc.  
  The number has no intermediate values between whole numbers.

- **Sensor Readings (Continuous):**  
  A humidity sensor might report 67.2%, 67.21%, 67.214% etc.  
  These values lie on a continuous scale and can be infinitely granular.

Some other examples:
- **Discrete**: Number of cars in a car park, test scores
- **Continuous**: Blood pressure, time taken to complete a task, weight

### PMF vs PDF: Key Differences


| Feature                    | PMF (Probability Mass Function)                  | PDF (Probability Density Function)                 |
|----------------------------|--------------------------------------------------|----------------------------------------------------|
| Applicable to              | Discrete random variables                        | Continuous random variables                        |
| Function type              | Maps outcomes to exact probabilities             | Maps outcomes to *density*, not probability        |
| Sum/Integral               | Total sum of all probabilities = 1               | Total area under the curve = 1                     |
| Example | $\mathrm{P(X=3) = 0.2}$| $\mathrm{P(X=3) = 0}$; but area under $f(x)$ around $x=3 > 0$     |
| Graph behaviour            | Spikes (bars at discrete points)                 | Smooth curve                                       |

# **Uniform Distributions**


A uniform distribution is one in which all outcomes are equally likely within a specified range.

* **Discrete Uniform**: Equal probability over a finite set of values (e.g., die roll outcomes).
* **Continuous Uniform**: Equal likelihood across a continuous interval (e.g., any number between 0 and 1).




Python Libraries Used

- `random`: Built-in Python module for generating random numbers (discrete or continuous).
- `matplotlib.pyplot`: Used to create simple plots to visualise distributions.
- `scipy.stats`: Part of SciPy library, provides pre-defined probability distributions for both discrete and continuous cases.


```python
import random     # built-in module for random number generation
import matplotlib.pyplot as plt   # matplotlib provides plotting functionalities
```

```python
# Using the random module

# random float number between 0.0 and 1.0
print(random.random())

# random integer between a and b, inclusive
print(random.randint(1, 10))

# random float number between a and b, inclusive
print(random.uniform(1.0, 5.0))
```

### Discrete Uniform Distribution : Simulating a Fair Die Roll

```python
# Simulate 100 die rolls and store in a list

die_rolls = [random.randint(1, 6) for x in range(100000)]
#die_rolls
```

```python
# @title Pre-defined function to plot experimental uniform distribution
def plot_discrete(outcome_list):
  # Plot
  plt.figure(figsize=(6, 4))
  plt.hist(outcome_list, bins=range(1, 8), align='left', rwidth=0.8, color='skyblue', edgecolor='black')
  plt.title("Discrete Uniform Distribution")
  plt.xlabel("Outcome")
  plt.ylabel("Frequency")
  plt.grid(axis='y', linestyle='--', alpha=0.6)
  plt.xticks(range(1, 7))
  plt.show()
```

```python
# To call the function, pass the outcome list as an argument
plot_discrete(die_rolls)
```

### Continuous Uniform Distribution

```python
# Simulate 100 values uniformly drawn from [0, 1]
uniform_continuous = [random.uniform(0, 1) for i in range(100000)]
```

```python
# @title Pre-defined function to plot experimental uniform distribution
def plot_continuous(outcome_list):
  # Plot histogram
  plt.figure(figsize=(6, 4))
  plt.hist(outcome_list, bins=20, density=True, color='lightgreen', edgecolor='black')
  plt.title("Continuous Uniform Distribution: U(0,1)")
  plt.xlabel("Value")
  plt.ylabel("Density")
  plt.grid(axis='y', linestyle='--', alpha=0.6)
  plt.show()
```

# **Bernoulli & Binomial Distributions**

### Bernoulli distribution
A Bernoulli distribution models a single trial with two possible outcomes:

**Success (1)** with probability $p$ and **Failure (0)** with probability $1 - p$

**Example**

- Suppose an email has a 30% chance of getting a click.
- One trial = one email → Outcomes: 0 (no click), 1 (click).


```python
from scipy.stats import bernoulli
```

```python
# Define a Bernoulli distribution with success probability p using bernoulli function from scipy
p = 0.3
bernoulli_dist = bernoulli(p)
```

```python
# PMF: P(X = 0) and P(X = 1)
x = [0, 1] # Define the possible outcomes for a Bernoulli trial (0 for failure, 1 for success)
pmf_vals = bernoulli_dist.pmf(x) # Calculate the probability for each outcome in x
```

```python
# @title Pre-defined function to plot bernoulli distribution

# Creating a function to plot bernoulli distribution from success probability, outcomes list, and PMF values
def plot_bern(pX, X, pmf):
  # Plot
  plt.figure(figsize=(4, 3))
  plt.bar(X, pmf, color='orange', edgecolor='black')
  plt.title(f"Bernoulli Distribution (p = {pX})")
  plt.xlabel("Outcome (0 = Fail, 1 = Success)")
  plt.ylabel("Probability")
  plt.xticks([0, 1])
  plt.ylim(0, 1)
  plt.grid(axis='y', linestyle='--', alpha=0.6)
  plt.show()
```

```python
plot_bern(p, x, pmf_vals)
```

### Binomial Distribution
A binomial distribution models the number of successes in $n$ independent Bernoulli trials.

**Example**

- Count of total clicks (successes) across 20 emails is a Binomial random variable
- Each email interaction is a Bernoulli trial

The binomial probability for $k$ success out of $n$ trials is given as:
$$P(X = k) = \binom{n}{k} p^k (1-p)^{n-k}$$

```python
from scipy.stats import binom
```

```python
# Parameters
n = 10      # number of emails sent
p = 0.5     # probability of a click on each email


# Define distribution
binom_dist = binom(n, p)

# Possible number of successes: 0 to n
x = list(range(n + 1))

# PMF
pmf_vals = binom_dist.pmf(x)
```

```python
# @title Pre-defined function to plot binomial distribution

# Creating a function to plot binomial distribution from success probability, list of possible successes, and PMF values
def plot_binom(n, pX, x_bin, pmf_bin):
  plt.figure(figsize=(8, 4))
  plt.bar(x, pmf_vals, color='skyblue', edgecolor='black')
  plt.title(f"Binomial Distribution: n={n}, p={p}")
  plt.xlabel("Number of Clicks (Successes)")
  plt.ylabel("Probability")
  plt.grid(axis='y', linestyle='--', alpha=0.6)
  plt.show()
```

```python
plot_binom(n, p, x, pmf_vals)
```

# **Normal Distribution & Z-Scores**

The Normal Distribution is a continuous probability distribution that is bell-shaped and symmetric around its mean ($μ$).

**Scenario: Human Heights**

Suppose the heights of adult women in a population are **normally distributed** with:

- Mean ($μ$) = 165 cm  
- Standard deviation ($σ$) = 6 cm

There would be more women around the average height. As we move further (higher or lower) from the mean, the number of women would keep decreasing.
<br><br>

We use this scenario to explore:
- Normal distribution
- Interval probabilities
- Z-scores and their interpretation


### `scipy.stats.norm`

- `scipy.stats.norm` represents the **Normal (Gaussian) distribution**.
- It is a continuous distribution characterised by:
  - `loc`: the mean ($μ$)
  - `scale`: the standard deviation ($σ$)

Common methods:
- `.pdf(x)`: Probability Density Function
- `.cdf(x)`: Cumulative Distribution Function
- `.ppf(q)`: Percent Point Function (inverse of CDF)

```python
from scipy.stats import norm

# Define parameters
mu = 165 #mean
sigma = 6 #sd

# Generate synthetic height data using the inverse CDF
data = [norm.ppf(random.random(), loc=mu, scale=sigma) for l in range(1000)]
```

```python
# @title Predefined code to plot histogram of the data
plt.figure(figsize=(10, 5))
count, bins, ignored = plt.hist(data, bins=30, density=True, alpha=0.6, color='skyblue', edgecolor='black', label='Simulated Heights')

# Display the theoretical PDF
x = [i for i in range(140, 190)]
pdf_vals = norm.pdf(x, loc=mu, scale=sigma)
plt.plot(x, pdf_vals, 'r--', label='Normal PDF (μ=165, σ=6)')

plt.title("Simulated Height Data with Normal PDF")
plt.xlabel("Height (cm)")
plt.ylabel("Density")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
